Para hablar o referirnos a insert de avl, tenemos que analizar el significado de la estructura avl en general, que se define como un arbol binario que se autobalancea cumpliendo con que cada sub arbol tenga una altura de 1 o 0

Dado esto la altura para un arbol binario es log base 2 de n 

entonces en terminos de busqueda de posicion para insertar la insercion deberia ser del mismo tamaño de la altura, log base 2 de n

pero dentro de la insercion esta el proceso de autobalanceo que hace swap de los nodos con el fin de volver a balancear el arbol, tomando en consideracion que el contenido de estos en constante el orden de la insercion por tanto es log base 2 de n
